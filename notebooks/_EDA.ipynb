### Análisis de la Destrucción de Infraestructura de Producción de Drogas Ilícitas por Departamento

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("..\\datasets\\processed\\drop_columns.csv")


df_grouped = df.groupby("DEPARTAMENTO")["CANTIDAD"].sum().reset_index()

# df_10datos=df_grouped.head(10)   <------------------------------------------------

ax=df_grouped.plot(x="DEPARTAMENTO", y="CANTIDAD", kind="bar", color="skyblue", legend=False)

plt.xlabel("Departamento")
plt.ylabel("Cantidad")
plt.title("Cantidad de destrucción de infraestructura \nde producción de drogas ilícitas por departamento")
plt.xticks(rotation=90)

# fig = ax.get_figure()
# fig.savefig("..\\media\\reporte.pdf", dpi=300, bbox_inches="tight")
# fig.savefig("..\\media\\reporte.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Verificación por departamento a selección por curiosidad

```python
total_antioquia = df.loc[df["DEPARTAMENTO"] == "HUILA", "CANTIDAD"].sum()

print("Total de unidades en Antioquia:", total_antioquia)
```

### Filtrado por año

```python

df["FECHA HECHO"] = pd.to_datetime(df["FECHA HECHO"])

df["AÑO"] = df["FECHA HECHO"].dt.year

df_grouped = df.groupby("AÑO")["CANTIDAD"].sum().reset_index()

df_grouped.plot(x="AÑO", y="CANTIDAD", kind="bar", color="skyblue", legend=False)
plt.xticks(rotation=90)  

plt.show()

# df_grouped = df_grouped.sort_values("AÑO")

print(df_grouped)
```
